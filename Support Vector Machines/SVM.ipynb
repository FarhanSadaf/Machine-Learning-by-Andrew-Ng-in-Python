```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
from sklearn.svm import SVC
```

### Dataset 1

```python
mat = loadmat("..\datasets\ex6data1.mat")
X = mat["X"]
y = mat["y"]
print(X.shape, y.shape)
```

```python
# Plotting the data
pos, neg = (y == 1).ravel(), (y == 0).ravel()
plt.scatter(X[pos, 0], X[pos, 1], c='g', marker='+', s=50)
plt.scatter(X[neg, 0], X[neg, 1], c='r', marker='.', s=50)
```

#### Linear Kernel

##### C = 1

```python
clf0 = SVC(kernel='linear')
clf0.fit(X, y.ravel())
```

```python
# Decision Boundary
u = np.arange(0, 4.5, 0.1)
v = np.arange(1, 5, 0.1)
z = np.zeros((len(u), len(v)))

for i in range(len(u)):
    for j in range(len(v)):
        z[i, j] = clf0.predict(np.array([u[i], v[j]]).reshape(1, -1))

pos, neg = (y == 1).ravel(), (y == 0).ravel()
plt.scatter(X[pos, 0], X[pos, 1], c='g', marker='+', s=50)
plt.scatter(X[neg, 0], X[neg, 1], c='r', marker='.', s=50)
plt.contour(u, v, z.T, 0)
```

##### C = 100

```python
clf1 = SVC(kernel='linear', C=100)
clf1.fit(X, y.ravel())
```

```python
# Decision Boundary
u = np.arange(0, 4.5, 0.1)
v = np.arange(1, 5, 0.1)
z = np.zeros((len(u), len(v)))

for i in range(len(u)):
    for j in range(len(v)):
        z[i, j] = clf1.predict(np.array([u[i], v[j]]).reshape(1, -1))

pos, neg = (y == 1).ravel(), (y == 0).ravel()
plt.scatter(X[pos, 0], X[pos, 1], c='g', marker='+', s=50)
plt.scatter(X[neg, 0], X[neg, 1], c='r', marker='.', s=50)
plt.contour(u, v, z.T, 0)
```

#### Gaussian Kernel

```python
clf2 = SVC(kernel='rbf')
clf2.fit(X, y.ravel())
```

```python
# Decision Boundary
u = np.arange(0, 4.5, 0.1)
v = np.arange(1, 5, 0.1)
z = np.zeros((len(u), len(v)))

for i in range(len(u)):
    for j in range(len(v)):
        z[i, j] = clf2.predict(np.array([u[i], v[j]]).reshape(1, -1))

pos, neg = (y == 1).ravel(), (y == 0).ravel()
plt.scatter(X[pos, 0], X[pos, 1], c='g', marker='+', s=50)
plt.scatter(X[neg, 0], X[neg, 1], c='r', marker='.', s=50)
plt.contour(u, v, z.T, 0)
```

### Dataset 2

```python
mat2 = loadmat("..\datasets\ex6data2.mat")
X2 = mat2["X"]
y2 = mat2["y"]
print(X2.shape, y2.shape)
```

```python
# Plotting the data
pos, neg = (y2 == 1).ravel(), (y2 == 0).ravel()
plt.scatter(X2[pos, 0], X2[pos, 1], c='g', marker='+')
plt.scatter(X2[neg, 0], X2[neg, 1], c='r', marker='.')
```

<img src="math 1.png" />
$$\gamma = \frac{1}{2\sigma^2}$$

##### gamma = 'scale'

```python
clf3 = SVC(kernel='rbf')
clf3.fit(X2, y2.ravel())
```

```python
# Decision Boundary
u = np.arange(0, 1, 0.01)
v = np.arange(0.4, 1, 0.01)
z = np.zeros((len(u), len(v)))

for i in range(len(u)):
    for j in range(len(v)):
        z[i, j] = clf3.predict(np.array([u[i], v[j]]).reshape(1, -1))

pos, neg = (y2 == 1).ravel(), (y2 == 0).ravel()
plt.scatter(X2[pos, 0], X2[pos, 1], c='g', marker='+')
plt.scatter(X2[neg, 0], X2[neg, 1], c='r', marker='.')
plt.contour(u, v, z.T, 0)
```

##### gamma = float
$$\sigma = 0.1$$
$$\gamma = \frac{1}{2 x 0.1^2} = 50$$

```python
clf4 = SVC(kernel='rbf', gamma=50)
clf4.fit(X2, y2.ravel())
```

```python
# Decision Boundary
u = np.arange(0, 1, 0.01)
v = np.arange(0.4, 1, 0.01)
z = np.zeros((len(u), len(v)))

for i in range(len(u)):
    for j in range(len(v)):
        z[i, j] = clf4.predict(np.array([u[i], v[j]]).reshape(1, -1))

pos, neg = (y2 == 1).ravel(), (y2 == 0).ravel()
plt.scatter(X2[pos, 0], X2[pos, 1], c='g', marker='+')
plt.scatter(X2[neg, 0], X2[neg, 1], c='r', marker='.')
plt.contour(u, v, z.T, 0)
```

### Dataset 3

```python
mat3 = loadmat("..\datasets\ex6data3.mat")
X3_train = mat3["X"]
y3_train = mat3["y"]
X3_val = mat3["Xval"]
y3_val = mat3["yval"]
print(X3_train.shape, y3_train.shape)
print(X3_val.shape, y3_val.shape)
```

```python
# Plotting the data
pos, neg = (y3_train == 1).ravel(), (y3_train == 0).ravel()
plt.scatter(X3_train[pos, 0], X3_train[pos, 1], c='g', marker='+', s=50)
plt.scatter(X3_train[neg, 0], X3_train[neg, 1], c='r', marker='.', s=50)
```

##### Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV

parameters = {
    'C': np.logspace(-3, 2, 10),
    'gamma': np.logspace(-3, 2, 10),
    'kernel': ['rbf', 'linear']
}

clf = GridSearchCV(SVC(), parameters, cv=5)
clf.fit(X3_val, y3_val.ravel())
print(clf.best_score_)
print(clf.best_params_)
```

```python
clf5 = clf.best_estimator_
clf5.fit(X3_train, y3_train.ravel())
```

```python
# Decision Boundary
u = np.arange(-0.7, 0.3, 0.01)
v = np.arange(-0.7, 0.7, 0.01)
z = np.zeros((len(u), len(v)))

for i in range(len(u)):
    for j in range(len(v)):
        z[i, j] = clf5.predict(np.array([u[i], v[j]]).reshape(1, -1))

pos, neg = (y3_train == 1).ravel(), (y3_train == 0).ravel()
plt.scatter(X3_train[pos, 0], X3_train[pos, 1], c='g', marker='+', s=50)
plt.scatter(X3_train[neg, 0], X3_train[neg, 1], c='r', marker='.', s=50)
plt.contour(u, v, z.T, 0)
```
